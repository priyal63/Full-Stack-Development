# Notice Board

```python
%%writefile 6_notice_board.py
import streamlit as st
from datetime import date
st.set_page_config(page_title="Notice Board",page_icon="‚öæ",layout="centered")
st.title("Notice Board")

st.sidebar.header("Filter Notices")
selected=st.sidebar.selectbox("Notice Board",["All","Exams","Workshops","Internships"])
show_past=st.sidebar.checkbox("Show past Notice",value=True)

notices=[
    {"title":"T4 Exam Scedule","category":"Exams","date":date(2026,1,1)},
    {"title":"Python Workshop","category":"Workshops","date":date(2026,1,5)},
    {"title":"Internship Orientation","category":"Internships","date":date(2026,1,3)},
]

st.header("Notices")
col1,col2=st.columns([1,2])
with col1:
    st.subheader("Filter Applied")
    st.write(f"**Category:** {selected}")
    st.write(f"Include Past Notice: {show_past}")
    
with col2:
    st.subheader("Information")
    st.text("Below are notices after filter")
    
for notice in notices:
    if selected !="All" and notice["category"] != selected:
        continue
    with st.expander(f"{notice["title"]} {notice["category"]}"):
        st.write(f"**Date** {notice["date"]}")
        st.write("Notice Details")
        
# python -m streamlit run 6_notice_board.py
```

# Date,Time & File Uploader

```python
%%writefile 7_date_time_file.py
import streamlit as st
from datetime import date,time
st.set_page_config(page_title="Date,Time & File Uploader",page_icon="‚öæ",layout="centered")
st.title("Date,Time & File Uploader demo")

exam_date=st.date_input("Select Exam Date: ",value=date.today())
start_time=st.time_input("Exam Start Time: ",value=time(9,0))

upload_file=st.file_uploader("Upload CSV file",type=["csv"])
st.write(f"Selected Exam Date: {exam_date}")
st.write(f"Exam start time: {start_time}")

if upload_file is not None:
    st.success("File Uploaded Successfully")
    st.write("File Name:  ",upload_file.name)
    st.write("File type:  ",upload_file.type)
```

# Buttons & Download Button

```python
%%writefile 8_button_demo.py
import streamlit as st
import pandas as pd
st.set_page_config(page_title="Buttons & Download Button",page_icon="‚öæ",layout="centered")
st.title("Buttons & Download Button demo")

if st.button("Click to Generate sample Data"):
    df=pd.DataFrame({
        "Enrollment Number ":[1,2,3,4,5],
        "Marks ":[78,85,69,92,89]
    })
    st.write("generated Data")
    st.dataframe(df)
    csv=df.to_csv(index=False).encode("utf-8")
    st.download_button(label="Download as CSV",data=csv,file_name="sample.csv",mime="text/csv")
```

```python
%%writefile 9_display_data.py
import streamlit as st
import pandas as pd
st.set_page_config(page_title="Dataframe,Table,JSON",page_icon="‚öæ",layout="centered")
st.title("Dataframe,Table,JSON demo")

data={
    "Student":["A","B","C","D"],
    "Marks":[85,90,98,45],
    "Passed":[True,True,True,False]
}

df=pd.DataFrame(data)
st.subheader("sr.dataframe(Interactive)")
st.dataframe(df)

st.subheader("st.table(static)")
st.table(df)

st.subheader("st.json(Structured JSON)")
st.json(data)
```

# Status element 

```python
%%writefile 11_status_demo.py
import streamlit as st
import time

st.set_page_config(page_title=" Status element",page_icon="‚öæ",layout="centered")
st.title(" Status element demo")

st.success("This is success Message")
st.warning("This is warning Message")
st.error("This is error Message")
st.info("This is info Message")

st.write("---")

st.subheader("Progress & Spinner Example")
if st.button("Start Long Tast"):
    progress=st.progress(0)
    with st.spinner("Progress......."):
        for i in range(100):
            time.sleep(0.03)
            progress.progress(i+1)
        st.success("Task Complated")
```

# Basic Matplotlib Charts

```python
%%writefile 12_matplotlib_basic.py
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

st.set_page_config(page_title="Basic Matplotlib Charts",page_icon="‚öæ",layout="centered")
st.title(" Basic Matplotlib Charts")

x=np.arange(1,11)
y=np.random.randint(50,100,size=10)

#-----------
# line Chart
#-----------

st.subheader("Line Chart (matplotlib)")
plt.figure(figsize=(6,4))
plt.plot(x,y,marker="o")
plt.xlabel("student index")
plt.ylabel("Marks")
plt.title("Marks of 10 student")
st.pyplot(plt)

plt.clf()

#-----------
# Bar Chart
#-----------

st.subheader("Bar Chart (matplotlib)")
plt.figure(figsize=(6,4))
plt.bar(x,y)
plt.xlabel("student index")
plt.ylabel("Marks")
plt.title("Marks of 10 student")
st.pyplot(plt)
```
